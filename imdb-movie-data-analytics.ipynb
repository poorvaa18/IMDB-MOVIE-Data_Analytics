```python
#10. In Which Year There Was The Highest Voting?
df.Votes.sort_values()
df.Votes.max()
df.groupby('Year')['Votes'].mean().idxmax() #idxmax() will return the maximum year index wrt to mean of the Votes
#max() will return the maximum value whereas s. idxmax() would return the index of the maximal value.

##data.groupby('Year')['Votes'] will make series of votes for each year
#data.groupby('Year')['Votes'].mean() will make series of mean votes for each year
#data.groupby('Year')['Votes'].mean().idxmax() will give the year with max mean votes
```

```python
#11. In Which Year There Was The Highest Average Revenue?
df.groupby('Year')['Revenue (Millions)'].mean().idxmax()
```

```python
#12. Find The Average Rating For Each Director
df.groupby('Director')['Rating'].mean()
```

```python
#13. Display Top 10 Lengthy Movies Title and Runtime
df.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']]
                    
```

```python
#14. Display Number of Movies Per Year
df['Year'].value_counts() #value_counts() it returns a series containing count(frequent occuring) of unique values
```

```python
#15. Find Most Popular Movie Title (Highest Revenue)
a=df['Revenue (Millions)'].max() #it will return the max revenue 
df[df['Revenue (Millions)']==a]['Title']
```

```python
#16. Display Top 10 Highest Rated Movie Titles And its Directors
c=df.nlargest(10, 'Rating')
#d=c.drop(columns=['Rank','Genre','Description','Actors','Year','Rating','Runtime (Minutes)', 'Votes','Revenue (Millions)','Metascore'])
d=c[['Title','Director']]
d
```

```python
#17. Display Top 10 Highest Revenue Movie Titles
topRevenue=df.nlargest(10,'Revenue (Millions)')[['Title','Revenue (Millions)']].set_index('Title')
topRevenue
sns.barplot(x='Revenue (Millions)',y=topRevenue.index,data=topRevenue)
```

```python
#18.  Find Average Rating of Movies Year Wise
df.groupby('Year')['Rating'].mean()
```

```python
#19. Does Rating Affect The Revenue?
sns.scatterplot(x='Rating',y='Revenue (Millions)',data=df)
#conclusion=Revenue is not dependent on the rating
```

```python
#20. Classify Movies Based on Ratings [Excellent, Good, and Average]
def Classify(rating):
    if rating>7:
        return 'Excellent'
    elif rating<7 and rating>=5:
        return "Good"
    else:
        return "Average"
    
```

```python
df['Rating'].apply(Classify)
```

```python
#21. Count Number of Action Movies
count=df[df['Genre'].str.contains('Action')] #Using “contains” to Find a Substring in a Pandas DataFrame
#The contains method in Pandas allows you to search a column for a specific substring.
#The contains method returns boolean values for the Series with True for if the original Series value contains the substring and False if not.
len(count)
```

```python
#22. Find Unique Values From Genre 
l1 =[]
for i in df['Genre']:
    l1.append(i.split(','))

l1

new=[]
for i in l1:
    for j in i: # j is seperating the list of list created in l1
        new.append(j)
new
```
